```python

orange_data_clean =  df[(df['age_bracket'] != '-') & (df['tier'] != '-') & (df['gender'] != '-') & (df['oPhone'] != '-') & (df['oWatch'] != '-') & (df['oPods'] != '-') & (df['oTV'] != '-') & (df['date'] != '-') & (df['card'] != '-')]
orange_data_clean
```

```python
orange_data_clean['oPhone']= orange_data_clean['oPhone'].astype(int)
orange_data_clean['oWatch']= orange_data_clean['oWatch'].astype(int)
orange_data_clean['oPods']= orange_data_clean['oPods'].astype(int)
orange_data_clean['oTV']= orange_data_clean['oTV'].astype(int)
```

```python
orange_data_clean.dtypes
```

```python
cols = ['oPhone', 'oWatch', 'oPods', 'oTV']
orange_data_clean['owns_multiple_products'] = orange_data_clean['oPhone'] + orange_data_clean['oWatch'] + orange_data_clean['oPods'] + orange_data_clean['oTV']

```

```python
orange_data_clean['owns_multiple_products'] = orange_data_clean['owns_multiple_products'].apply(lambda x: 1 if x >= 2 else 0 )
orange_data_clean['owns_multiple_products'] = orange_data_clean['owns_multiple_products'].astype(int)
orange_data_clean.dtypes
```
